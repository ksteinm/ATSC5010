### Nested loop example

This code provides an example of a nested loop.

Suppose you would like to know the density at every pressure level and temperature, but the pressure and temperature arrays are different sizes.

### Libraries

First you must load the necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Constants

Next, state any constants you need. 

```python
R_star = 8.314
```

### Creating temperature and pressure arrays

Create a 1300 element temperature array from -15 to 50.
Create a 2000 element pressure array from 1000 to 0.

```python
Temperature = np.linspace(-15, 50, 1300)
Temp_K = Temperature + 273.15

Pressure = np.linspace(0, 1000, 2000)
Press_Pa = Pressure * 1000
```

### Create a density variable

This new array should be a 2-D array with one dimension the size of the temperature array and the other dimension the size of the pressure array

```python
rho = np.empty((len(Press_Pa), len(Temp_K)))
rho[:] = np.nan
```

### Creating nested loop

This loop will go through both the temperature and pressure arrays to calculate density for each temperature and pressure point.

```python
for i in range(len(Press_Pa)):
    for j in range(len(Temp_K)):
        rho[i, j] = Press_Pa[i] / (R_star * Temp_K[j])
```

### Creating a contour levels array

This array is used to tell the function which lines it should plot

```python
rho_contour = np.arange(0, 500.1, 10)
```

### Plot

Create a contour plot for density

```python
plt.figure(figsize=(8, 9), facecolor = 'white')
density = plt.contour(Temperature, Pressure, rho, rho_contour)
plt.clabel(density, np.arange(0, 500, 50), inline=1)
plt.yscale('log')
plt.gca().invert_yaxis()
plt.ylim(1000, 100)
plt.yticks(np.arange(100, 1000.1, 100), ['100', '200', '300', '400', '500', '600', '700', '800', '900', '1000'])
plt.xlabel('Temperature ($^o$C)')
plt.ylabel('Pressure (mb)')
```
